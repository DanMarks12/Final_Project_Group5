```python
import pandas as pd
esg_df = pd.read_csv('ESG Scores .csv', index_col=0)
print(esg_df)
```

```python
stock_df = pd.read_csv('top_5_v1_df.csv', index_col=0)
print(stock_df)
```

```python
dummy_vars = ['MSCI ESG 2017', 'MSCI ESG 2018', 'MSCI ESG 2019', 'MSCI ESG 2020', 'MSCI ESG 2021', 'Sustainalytics', 'Sustainalytics Exposure', 'Sustainalytics Management']
dummy_df = pd.get_dummies(esg_df[dummy_vars])
```

```python
other_vars = list(set(esg_df.columns.tolist()) - set(dummy_vars))
esg_df = pd.concat([esg_df[other_vars], dummy_df], axis=1)
esg_df
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
corr_matrix = esg_df.corr()
```

```python
esg_df.dtypes
```

```python
sn.heatmap(corr_matrix, annot=False, cmap='Blues')
plt.show()
```
