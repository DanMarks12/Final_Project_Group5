```python
# Load stock data
stock_info = "Resources/tech_stocks.csv"
stock_info_df = pd.read_csv(stock_info)
stock_info_df.head(10)
```

```python
# rename columns 
stock_info_df.rename(columns = {"Symbol": "ticker",
                     "Name":"c_name",
                    "Last Sale": "last_sale",
                    "Market Cap": "market_cap",
                    "IPO Year": "ipo_year",
                    "Sector": "sector",
                    "Industry": "industry"},
                                 inplace = True)
```

```python
# pandas drop columns using list of column names
new_stock_info = stock_info_df.drop(['% Change', 'Volume', 'Country', 'Net Change'], axis=1)
new_stock_info.head(10)
```

```python
# Remove rows that have at least 1 null value.
df_stocks = new_stock_info.dropna()
df_stocks.head(10)
```
