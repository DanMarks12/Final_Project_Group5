## Getting started with ARIMA

##### First ARIMA model without splitting data

```python
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.arima_model import ARIMA
```

```python
#check length of df
len(mrnadf)
```

```python
#making arima model with pdq as 1,1,1 without fitting data

model = ARIMA(mrnadf['Close'].fillna(value=mrnadf['Close'].mean()),order=(6,1,6))
model_fit= model.fit()
model_fit.params
result= model.fit(disp=False)
```

```python
result.summary()
```

```python
#use ARIMA predict function to apply predictions to variable = y_predict
y_predict = model_fit.predict(typ='levels')
print(y_predict)
```

```python
mrnadf['forecast']= model_fit.predict(start=468,typ='levels')
mrnadf[['Close', 'forecast']].plot(figsize=(16,8))
#save fig
plt.savefig("notfit_ARIMA_MSFT.png",)

plt.show()
```

```python
#check MSE score
nosplit_mse = mean_squared_error(mrnadf['Close'][1:len(mrnadf)],y_predict)

mean_squared_error(mrnadf['Close'][1:len(mrnadf)],y_predict)
```

###### ARIMA with split of y_train and y_test 

```python
#get close data alone

mrnadf = mrnadf[['Close']].copy()
mrnadf.info
```

```python
#split data into test and split at .80 (of )
y_train = mrnadf.iloc[:468]
y_test = mrnadf.iloc[468:]
```

```python
y_train.tail()
```

```python
y_test.head()
```

```python
#plot train and test
plt.figure(figsize=(16,8))
plt.plot(y_train,label='Train')
plt.plot(y_test,label='Actual')
plt.ylabel("Stock prices")
plt.title('MRNA Stock Prices',fontsize=20)
plt.show()
```

```python
print("start:",y_test.index.min())
print("end:",y_test.index.max())
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
arima = ARIMA(y_train,order=(6,1,3)).fit(disp=False)
y_arima = arima.predict(468,586,typ='levels')
y_arima.index = y_test.index
```

```python
#Check MSE of model
arima_mse = mse(y_arima,y_test)

mse(y_arima,y_test)
```

```python
#Plot ARIMA model

plt.figure(figsize=(16,8))
plt.plot(y_train,label='Train')
plt.plot(y_test,label='Actual')
plt.plot(y_arima,label='ARIMA Predict')
plt.ylabel("Stock prices")
plt.title('ARIMA Model',fontsize=20)
#save fig
plt.savefig("ARIMA_MSFT.png",)

plt.show()
```

## Checking ARMA Model
